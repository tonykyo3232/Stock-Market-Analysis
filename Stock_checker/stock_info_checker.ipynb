## Linear Regression in Mechine Learning

```python
# use Apple as example
compamny_name = 'Apple'
df = container[0][1]
```

```python
df
```

```python
# create a python list of feature names that would like to pick from the dataset:
feature_cols = ['Open', 'High','Low', 'Volume']

# use the above list to select the features from the original DataFrame
X = df[feature_cols]

y = df['Close']
```

```python
X
```

### Data Split

```python
from sklearn.model_selection import train_test_split
# Linear regression Model for stock prediction 
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.15 , shuffle=False, random_state = 0)
```

```python
# let's check if total observation makes sense
print(train_X.shape )
print(test_X.shape)
print(train_y.shape)
print(test_y.shape)
```

### linear regression model

General format for linear regression with multiple variables:

$y = \theta_0 + \theta_1x_1 + \theta_2x_2 + ... + \theta_nx_n$

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion_matrix, accuracy_score

regression = LinearRegression()
regression.fit(train_X, train_y)

# printing Theta0 using attribute "intercept_":
print("regression intercept",regression.intercept_)

# printing [Theta1, Theta2, Theta3] using attribute "coef_":
print("regression coefficient",regression.coef_)
```

#### Thus, our predictive model will be: 
$$y = 4.73 +  -2.83 \times Open + 7.09 \times High +  5.55 \times Low + -5.75 \times Income $$

## Prediction

```python
predicted=regression.predict(test_X)
print(test_X.head())
```

## Model Evaluation

- In regression, the target is continous valued. So, we need to find the error as the average difference between the "predicted target value" and the "actual value".
- The most popular metric to quantify this difference (error) is **Root Mean Square Error** or **RMSE**:

$$RMSE = \sqrt{\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2}$$
- "yi" is the actual value. "yi_hat" is the predicted value.
- Notice that **RMSE** is error. So, unlike the accuracy for classifiers, **a lower value for RMSE is better!**

### R²: coefficient of determination
- This measure is represented as a value between 0.0 and 1.0, where a value of 1.0 indicates a perfect fit, and is thus a highly reliable model for future forecasts, while a value of 0.0 would indicate that the model fails to accurately model the data at all.
- An R2 of 0 means that the dependent variable cannot be predicted from the independent variable.
- An R2 of 1 means the dependent variable can be predicted without error from the independent variable.
- An R2 between 0 and 1 indicates the extent to which the dependent variable is predictable. An R2 of 0.10 means that 10 percent of the variance in Y is predictable from X; an R2 of 0.20 means that 20 percent is predictable; and so on.

```python
from sklearn import metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print('Mean Absolute Error (MAE):', mean_absolute_error(test_y, predicted))
print('Mean Squared Error (MSE) :', mean_squared_error(test_y, predicted))
print('Root Mean Squared Error (RMSE):', np.sqrt(mean_squared_error(test_y, predicted)))
print('Coefficient of determination:', r2_score(test_y, predicted))
```

```python
# alternative way to calculate the coefficient of determination R² 
regression_confidence = regression.score(test_X, test_y)
print("linear regression confidence: ", regression_confidence)
```
